```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
import tensorflow as tf
import os
```

```python
%load_ext tensorboard
```

```python
classifier = Sequential()
```

- `convolution`

```python
classifier.add(Convolution2D(
    32, 3, 3, input_shape=(64, 64, 3), activation='relu'))
```

- `pooling` 

```python
classifier.add(MaxPooling2D(pool_size=(2, 2)))
```

- `adding second convolution layer`

```python
classifier.add(Convolution2D(
    32, 3, 3, activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2, 2)))
```

- `flattening`

```python
classifier.add(Flatten())
```

- `full connection`

```python
# hidden layer
classifier.add(Dense(activation='relu', units=128))

# classifier.add(Dense(activation='relu', units=64))

# last layer / output layer
# sigmoid coz we have only 1 output, if the o/p was more than 2 use softmax
classifier.add(Dense(activation='sigmoid', units=1))
```

```python
classifier.compile(
    optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
tboard_log_dir = os.path.join("logs")

tensorboard_callback = tf.keras.callbacks.TensorBoard(
    log_dir=tboard_log_dir, histogram_freq=1)

# classifier.fit(x=X,
#                y=y,
#                epochs=10,
#                batch_size=10,
#                callbacks=[tensorboard_callback])
```

- `image augmentation technique to overcome over-fitting `

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train = train_datagen.flow_from_directory(
    'data/train/',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary')
```

```python
test = test_datagen.flow_from_directory(
    'data/test/',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary')
```

```python
train_steps = train.n // 32  # 2000/32
test_steps = test.n // 32  # 1000/32
print(train_steps, test_steps)
```

```python
classifier.fit(
    train,
    steps_per_epoch=train_steps,
    epochs=100,  # 322
    validation_data=test,
    validation_steps=test_steps,
    callbacks=[tensorboard_callback])  # callbacks=[tensorboard_callback]
```
