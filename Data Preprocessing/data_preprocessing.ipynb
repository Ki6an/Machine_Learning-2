```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('dataset/Data.csv')
df
```

```python
X = df.values[:-1, :-3]
X
```

```python
y = df.values[:-1, 3]
y
```

handling the missing values, changing the 'nan' value with mean of the respective column

```python
impute = SimpleImputer(missing_values=np.nan, strategy='mean')
impute = impute.fit(X[:, 1:3])
X[:, 1:3] = impute.transform (X[:, 1:3])
X
```

the names of the countries are encoded/ converted into numbers, but this doesn't work coz no country has greater value than other. 

```python
from sklearn.preprocessing import LabelEncoder

labelencoder_X = LabelEncoder()

X[:, 0] = labelencoder_X.fit_transform(X[:, 0])
X
```

```python
from sklearn.preprocessing import OneHotEncoder # depricated 

from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([('encoder', OneHotEncoder(), [0])],
                           remainder='passthrough')

X = np.array(ct.fit_transform(X))

X
```

now lets do the same for label i.e y

```python
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
y
```

test and train split 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(X_train.shape,
    X_test.shape,
    y_train.shape,
    y_test.shape)
```

feature scaling to get all the data on the same page 

```python
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)

X_train
X_test
```
